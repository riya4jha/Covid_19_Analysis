***Covid-19 Data Analysis Using Pandas***

**Importing Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot  as plt
```

**Read CSV File**

```python
df1=pd.read_csv('raw_data2.csv')
df2=pd.read_csv('raw_data4.csv')
df2
```

**Inspecting any data frame**

```python
df2.info()
```

**Displaying all column names**

```python
df1.columns
```

**Retaining Necessary Columns**

```python
df1=df1.loc[:,['Num Cases','Date Announced', 'Age Bracket', 'Gender', 'Detected City','Detected State','Current Status',]]
df1
```

```python
df2=df2.loc[:,['Num Cases','Date Announced', 'Age Bracket', 'Gender', 'Detected City','Detected State','Current Status',]]
df2

#iloc can also be used df1=df1.iloc[:,[0,1,2,3,4,5,6,20]]
```

**Merging both the Data Frames**

```python
df=df1.append([df2])
df
```

**Inspect Null Values in Each Column**

```python
df.isnull().sum().sort_values(ascending=False)/len(df)*100
#percent of values missing from actual data
```

**Inspect Null Values in each row**

```python
df.isnull().sum(axis=1).sort_values(ascending=False)
```

**Separating Date,Month and Year From Date Announced**

```python
DATE=df["Date Announced"].str.split('/',expand=True)
#expand makes 3 different columns and makes a df
DATE.columns=['Day','Month','Year']
DATE
```

**Concatenating Date with previous Data frame df1**

```python
df=pd.concat([df,DATE],axis=1)
df
```

```python
df.to_csv('Covid19India.csv')
```

**Total Number of cases Monthly**

```python
df.groupby('Month')['Num Cases'].sum()
```

```python
Data=df[df['Current Status']=="Hospitalized"].groupby('Month')['Num Cases'].sum()
Data
```

```python
Data.plot.bar()
plt.show()
```

**Total male/Female infected by Covid**

```python
df.groupby('Gender')['Num Cases'].sum()
```

**Age Group that is infected the Most**

```python
data=df.groupby('Age Bracket')['Num Cases'].sum().sort_values(ascending=False).head(10)
```

```python
df.groupby('Age Bracket')['Num Cases'].sum().max()
```

```python
data.plot.bar()
plt.show()
```

**Statewise Total Number Of Cases In India**

```python
state_data=df[df['Current Status']=='Hospitalized'].groupby('Detected State')['Num Cases'].sum().sort_values(ascending=False).head(10)
```

```python
state_data.plot.bar()
plt.show()
```

**Cases Daily**

```python
day=df[df['Current Status']=='Hospitalized'].groupby(['Month','Day'])[['Num Cases']].sum()
```

```python
day.unstack(level=0).plot(kind='bar',subplots=True,figsize=(10,10))
plt.show()
```

**Deaths In Each Country**

```python
df[df['Current Status']=='Deceased']['Num Cases'].sum()
```

```python
deaths=df[df['Current Status']=='Deceased'].groupby('Detected State')['Num Cases'].sum().sort_values(ascending=False).head(10)
deaths.plot.bar()
plt.show()
```
